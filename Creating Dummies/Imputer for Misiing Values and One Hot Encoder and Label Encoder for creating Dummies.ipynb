```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Missing data 28-08-2020.csv")
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values = np.nan,  
                        strategy ='mean') 
```

```python
imp = imp.fit(df[['Age', 'Salary']]) 
```

```python
df[['Age', 'Salary']] = imp.transform(df[['Age', 'Salary']])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer 
```

```python
cT = ColumnTransformer([('encoder', OneHotEncoder(), [3])],  remainder='passthrough')
```

```python
df =cT.fit_transform(df)
```

```python
df
```

```python
data = pd.DataFrame(df, columns=["Purchased", " Not Purchased", "Country", "Age", "Salary"])
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() 
```

```python
data['Country']= le.fit_transform(data['Country']) 
```
